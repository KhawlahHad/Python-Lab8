```python
#Product Class

class Product:
    
    def __init__(self, name, price, quantity):
        try:
            self.name = name
            self.price = float(price)
            self.quantity = int(quantity)
        except:
            raise TypeError("Invalid value")


    def update_quantity(self, quantity):
        try:
            self.quantity = int(quantity)
        except:
            raise TypeError("Only integers are allowed")

    
    def get_total_price(self):
        return self.quantity * self.price

    
    def display_product_details(self):
        print(f'Properties of the Product:\n Name: {self.name}\n Price: {self.price}\n \
              Quantity: {self.quantity}\n Total Price: {self.get_total_price()}\n')
      
```

```python
#Inventory Class

class Inventory:
    def __init__(self):
        self.inventory = {} 

    def addProduct(self, product):
        self.inventory[product.name] = product

    def updateProduct(self, product):
        try:
            self.inventory[product.name] = product
        except:
            print(f"Product {product.name} doesn't exist")
            
    def removeProduct(self, name):
        try:
            self.inventory.pop(name) 
        except:
            print(f"Product {name} doesn't exist")

    def totalCost(self):
        totalCost = 0
        for product in self.inventory.values():
            totalCost+= product.get_total_price()
        return totalCost
        
    def printInventory(self):
        print("Inventory:\n---------")
        for key in self.inventory.keys():
            print(f'Product: {key} \n Price: {self.inventory[key].price}, Quantity: {self.inventory[key].quantity}')
        print('\n')


            

            
        
```

```python
#Testing

try:
    p = Product("myProduct", 9, 10)
    p.update_quantity(90)
    p.display_product_details()

    p1 = Product("myProduct1", 9, 10)
    p2 = Product("myProduct2", 54, 12)
    p3 = Product("myProduct3", 54, 12)
    
    myInventory = Inventory()
    myInventory.addProduct(p1)
    myInventory.addProduct(p2)
    myInventory.addProduct(p3)

    myInventory.printInventory()  #print before change

    myInventory.updateProduct(p)
    myInventory.removeProduct("myProduct3")

    myInventory.printInventory() #print after change

    print(f'Total inventory cost: {myInventory.totalCost()}')


    
except Exception as e:
    print(f'Error message: {e}')
```
